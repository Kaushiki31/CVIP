## Importing required libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

## Reading the csv file

```python
data=pd.read_csv("breast-cancer.csv")
data.head(10)
```

```python
data.shape
```

```python
data.info()
```

## Checking the missing values

```python
data.isnull().sum()
```

## Checking the correlation matrix

```python
fig,ax=plt.subplots(figsize=(30,30))
corr=data.corr()
sns.heatmap(corr,annot=True,ax=ax)
```

## Applying Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
data['diagnosis']=le.fit_transform(data['diagnosis'])
data.head(10)
```

```python
data.info()
```

## Splitting our data into training and testing dataset



```python
from sklearn.model_selection import train_test_split
```

```python
x=data.drop(['id','diagnosis'],axis='columns')
y=data['diagnosis']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=42)
```

```python
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

## Training our model using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test) # Score is pretty good
```

```python
y_pred=model.predict(x_test)
```

## Using confusion matrix

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
plt.figure(figsize=(5,6))
sns.heatmap(cm,annot=True)
plt.xlabel('symptoms')
plt.ylabel('diagnosis')
plt.title('Breast cancer Prediction')
plt.show()
```
