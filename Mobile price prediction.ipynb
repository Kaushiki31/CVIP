## Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Reading the train dataset

```python
data=pd.read_csv('train.csv')
data.head(10)
```

## Checking the missing values

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data.columns
```

## Splitting the dataset into train and test dataset

```python
x=data.drop('price_range',axis='columns')
y=data['price_range']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

## Using Linear Regression to check model accuracy

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

## Using Logistic Regression to check model accuracy

```python
from sklearn.linear_model import LogisticRegression
model1=LogisticRegression()
```

```python
model1.fit(x_train,y_train)
```

```python
model1.score(x_test,y_test)
```

## Using Random Forest Classifier to check model accuracy

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
```

```python
rf.score(x_test,y_test)
```

## Using Decision Tree Classifier to check model accuracy

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
```

```python
dt.fit(x_train,y_train)
```

```python
dt.score(x_test,y_test)
```

## Using SVC to check model accuracy

```python
from sklearn.svm import SVC
model=SVC()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

### By checking we found that SVC gave the best score, so we will use SVC to predict our model

## Confusion Matrix

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
sns.heatmap(cm,annot=True)
plt.show()
```

## Classification Report

```python
from sklearn.metrics import classification_report
cr=classification_report(y_test,y_pred)
cr
```

## Plotting graph


```python
plt.scatter(y_test,y_pred,color='Red')
plt.plot(y_test,y_pred)
```

```python
sns.barplot(y_test,y_pred)
plt.xlabel('Actual Price Range')
plt.ylabel('Predicted Price Range')
plt.show()
```
