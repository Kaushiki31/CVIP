## Building Our Machine Learning Model

```python
df.head(20)
```

## Using Label Encoder


```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['country_txt'] = le.fit_transform(df['country_txt'])
df['region_txt'] = le.fit_transform(df['region_txt'])
df['attacktype1_txt'] = le.fit_transform(df['attacktype1_txt'])
df['targtype1_txt'] = le.fit_transform(df['targtype1_txt'])
```

```python
df.head(20)
```

```python
x = df[['country_txt','region_txt','attacktype1_txt','targtype1_txt','nkill']]
```

```python
x.head()
```

```python
y = df['success']
```

```python
y.head()
```

```python
x.shape
```

```python
y.shape
```

## splitting our model into training and testing dataset

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
xtrain.shape , ytrain.shape , xtest.shape , ytest.shape 
```

## Training our model using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtest, ytest) # Score not accurate enough
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
xtrain_new = scaler.fit_transform(xtrain)
xtest_new = scaler.transform(xtest)
```

```python
model.fit(xtrain_new,ytrain)
```

```python
model.score(xtest_new,ytest)
```

### Score not accurate enough, so we will be trying other method

## Training our model using Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth=10,random_state=42)
```

```python
clf.fit(xtrain,ytrain)
```

```python
clf.score(xtest,ytest) # Till now best score
```

## Confusion matrix

```python
y_pred=clf.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix=(ytest,y_pred)
cm
```
